```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['pdf.fonttype'] =42
import matplotlib.ticker as ticker
import matplotlib.patches as mpatches
```

```python
df_mus = pd.read_csv("MusicGenrePreference.csv")
df_mus.head()
```

```python
df_stress = pd.read_csv("FluSymptoms.csv")
df_stress.head()
```

```python
df_stress.shape
```

```python
df_stress.groupby('user_id')['open.stressed'].value_counts().head()
```

```python
df_new1 = df_stress.loc[df_stress['open.stressed'] == 1]
df_new1.head()
```

```python
df1 = df_new1.merge(df_mus, left_on= 'user_id', right_on='user.id')
df1
```

```python
popular = df1.groupby('user_id')['genre'].value_counts()
popular.groupby(level=0).nlargest(1).head()
```

```python
df_new2 = df_stress.loc[df_stress['open.stressed'] == 0]
df_new2.head()
```

```python
df2 = df_new2.merge(df_mus, left_on= 'user_id', right_on='user.id')
df2
```

```python
popular2 = df2.groupby('user_id')['genre'].value_counts()
popular2.groupby(level=0).nlargest(1)
```
