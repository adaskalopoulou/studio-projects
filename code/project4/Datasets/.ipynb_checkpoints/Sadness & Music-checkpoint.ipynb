```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['pdf.fonttype'] =42
import matplotlib.ticker as ticker
import matplotlib.patches as mpatches
```

```python
df_mus = pd.read_csv("MusicGenrePreference.csv")
df_mus.head()
```

```python
df_sad = pd.read_csv("FluSymptoms.csv")
df_sad.head()
```

```python
df_sad.shape
```

```python
df_sad.groupby('user_id')['sad.depressed'].value_counts().head()
```

```python
df_new1 = df_sad.loc[df_sad['sad.depressed'] == 1]
df_new1.head()
```

```python
df1 = df_new1.merge(df_mus, left_on= 'user_id', right_on='user.id')
df1.head()
```

```python
popular1 = df1.groupby('user_id')['genre'].value_counts()
popular1.groupby(level=0).nlargest(1)
```

```python
df_new2 = df_sad.loc[df_sad['sad.depressed'] == 1]
df_new2.head()
```

```python
df2 = df_new1.merge(df_mus, left_on= 'user_id', right_on='user.id')
df2.head()
```

```python
popular2 = df2.groupby('user_id')['genre'].value_counts()
popular2.groupby(level=0).nlargest(1)
```
