### checking the jokes with regex

```python
df_jokes = pd.read_csv("jester_items.csv")
pd.set_option('display.max_rows', 200)
df_jokes
```

```python
pd.options.display.max_colwidth = 1000000
```

```python
my_regex = r"\b doctor \b"
doctor = [joke for joke in df_jokes['Text'] if re.search(my_regex, joke,re.IGNORECASE)]
len(doctor)
```

```python
my_regex = r"\b no \b"
no = [joke for joke in df_jokes['Text'] if re.search(my_regex, joke,re.IGNORECASE)]
len(no)
```
