## Obtaining the data through scrapping

```python
import selenium
from selenium import webdriver
driver = webdriver.Chrome()
```

```python
def IDs(row):
    driver.get("http://www.travelwriting.amdigital.co.uk/Search/AdvancedSearch.aspx")
    textbox = driver.find_element_by_xpath('//*[@id="ctl11_ctl01_BooleanForm_ctl00_bs0_BooleanSearchTextBox"]')
    textbox.send_keys(row['IDs'])
    driver.find_element_by_xpath('//*[@id="SearchButton"]').click()
    driver.find_element_by_xpath('//*[@id="SearchResultRepeater_ctl02_ItemRow"]/td[2]/span[2]').click()
    try:
        Reference = driver.find_element_by_xpath('//*[@id="ReferenceRow"]/div[2]').text.strip()    
        Traveler = driver.find_element_by_xpath('//*[@id="TravellerRow"]/div[2]').text.strip()
        Title = driver.find_element_by_xpath('//*[@id="TitleRow"]/div[2]').text.strip()
        Date = driver.find_element_by_xpath('//*[@id="DateRow"]/div[2]').text.strip()
        Summary = driver.find_element_by_xpath('//*[@id="SummaryRow"]/div[2]').text.strip()
        Region = driver.find_element_by_xpath('//*[@id="RegionRow"]/div[2]').text.strip()
        Country = driver.find_element_by_xpath('//*[@id="CountryRow"]/div[2]').text.strip()
        Topics = driver.find_element_by_xpath('//*[@id="TopicsRow"]/div[2]').text.strip()

        return pd.Series({
            'Reference': Reference,
            'Traveler': Traveler,
            'Title': Title, 
            'Date': Date,
            'Summary': Summary,
            'Region': Region,
            'Country': Country,
            'Topics': Topics
        })
    except: 
        pass
         
driver = webdriver.Chrome()
df_abroad = df.apply(IDs,axis=1)
df_abroad
```

```python
def IDsUSA(row_usa):
    driver.get("http://www.travelwriting.amdigital.co.uk/Search/AdvancedSearch.aspx")
    textbox = driver.find_element_by_xpath('//*[@id="ctl11_ctl01_BooleanForm_ctl00_bs0_BooleanSearchTextBox"]')
    textbox.send_keys(row_usa['IDsUSA'])
    driver.find_element_by_xpath('//*[@id="SearchButton"]').click()
    driver.find_element_by_xpath('//*[@id="SearchResultRepeater_ctl02_ItemRow"]/td[2]/span[2]').click()
    try:
        Reference = driver.find_element_by_xpath('//*[@id="ReferenceRow"]/div[2]').text.strip()    
        Traveler = driver.find_element_by_xpath('//*[@id="TravellerRow"]/div[2]').text.strip()
        Title = driver.find_element_by_xpath('//*[@id="TitleRow"]/div[2]').text.strip()
        Date = driver.find_element_by_xpath('//*[@id="DateRow"]/div[2]').text.strip()
        Summary = driver.find_element_by_xpath('//*[@id="SummaryRow"]/div[2]').text.strip()
        Region = driver.find_element_by_xpath('//*[@id="RegionRow"]/div[2]').text.strip()
        Country = driver.find_element_by_xpath('//*[@id="CountryRow"]/div[2]').text.strip()
        Topics = driver.find_element_by_xpath('//*[@id="TopicsRow"]/div[2]').text.strip()

        return pd.Series({
            'Reference': Reference,
            'Traveler': Traveler,
            'Title': Title, 
            'Date': Date,
            'Summary': Summary,
            'Region': Region,
            'Country': Country,
            'Topics': Topics
        })
    except: 
        pass
         
driver = webdriver.Chrome()
df_usa = df_usa.apply(IDsUSA,axis=1)
df_usa
```

## Adding information about the writers/travelers

```python
traveler_list = []
all_countries = [df_usa, df_China, df_France, df_Philippines, df_India, df_Egypt, df_Italy]
for country in all_countries:
    myname = country['Country'].head(1)
    all_text_trav = []
    for trav in country.Traveler:
        all_text_trav.append(trav)

    all_text_trav = Counter(all_text_trav)    
    #print(myname.to_string(index=False))
    counter=0
    rows = {}
    for item in all_text_trav.most_common(40):
        counter=counter + 1
        item = {
            'name': myname.to_string(index=False),
            'traveler': item[0],
            'frequency' : item[1],
            'rank':counter    
    }
        traveler_list.append(item)
print(traveler_list)  
```

```python
df_trav = pd.DataFrame(traveler_list)
df_trav.to_csv("pr_traveler.csv", index=False)
df_trav
```

```python
import numpy as np
df_trav["article_trav"] = df_trav["traveler"] + " wrote " + df_trav["frequency"].map(str) + np.where(df_trav["frequency"]>1, ' pieces', ' piece') +" of travel writing in " + df_trav["name"]+"."
df_trav
```

```python
df_trav = df_trav.groupby('name')['article_trav'].apply(lambda x: "<p>%s</p>" % '</p><p> '.join(x)).reset_index(name='article_trav')
df_trav
df_trav.iloc[0]['article_trav']
```
