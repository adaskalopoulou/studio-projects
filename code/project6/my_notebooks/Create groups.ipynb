```python
df_usaornt = pd.read_csv("usaornot.csv")
df_usaornt
```

```python
df_usa = pd.read_csv('subset-usa.csv')
df_usa
```

```python
from collections import Counter
```

```python
my_group_list = []
all_groups = [df_usaornt, df_usa]
for group in all_groups:
    myname = group['Country'].head(1)
    all_text_group = []
    for topic in group.Topics:
        pieces = topic.split(", ")
        all_text_group.append(pieces)
    
    big_list_group = []
    for my_list in all_text_group:
        for my_elem in my_list:
            big_list_group.append(my_elem)

    big_list_group = Counter(big_list_group)    
    #print(myname.to_string(index=False))
    counter=0
    rows = {}
    for item in big_list_group.most_common(5):
        counter=counter + 1
        item = {
            'name': myname.to_string(index=False),
            'topic': item[0],
            'frequency' : item[1],
            'rank':counter    
    }
        my_group_list.append(item)
print(my_group_list)  
```

```python
df = pd.DataFrame(my_group_list)
df
```

```python
df.to_csv("pr_group.csv", index=False)
```
